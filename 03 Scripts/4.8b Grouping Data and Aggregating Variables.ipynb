```python
# Importing libraries

import pandas as pd
import numpy as np
import os
```

```python
# Path to IC folders
path = r"C:\Users\alabaro\OneDrive - DAI\Documents\Aug_2023 Instacart Basket Analysis"
```

```python
# importing orders_products_merged_v2
df_ords_prods_merged_v3 = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged_v3.pkl'))
```

```python
# define full dataset

ords_prods_merged = df_ords_prods_merged_v3
```

```python
# check output

ords_prods_merged.head(5)
```

```python
# group ords_prods_merged by user_id and find median of days_since_prior_order

ords_prods_merged['med_frequency'] = ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# check output

ords_prods_merged.head(5)
```

```python
# Set customer spending flag for non frequent customer

ords_prods_merged.loc[ords_prods_merged['med_frequency'] > 20, 'frequency_flag'] = 'Non frequent customer'
```

```python
# Set customer spending flag for regular customer

ords_prods_merged.loc[(ords_prods_merged['med_frequency'] <= 20) & (ords_prods_merged['med_frequency'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
# Set customer spending flag for frequent customer

ords_prods_merged.loc[ords_prods_merged['med_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# check output 

ords_prods_merged[['user_id', 'frequency_flag', 'order_number']].head(15)
```

```python
# check output

ords_prods_merged['frequency_flag'].value_counts(dropna=False)
```
