# 03. Create Flag for Customer loyalty

```python
# group ords_prods_merged by user_id and find max order_number

ords_prods_merged['max_order'] = ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# check ouput

ords_prods_merged.head(15)
```

```python
pd.options.display.max_rows = None
```

```python
# check output for first 100 rows

ords_prods_merged.head(100)
```

```python
# Set Loyal_customer flag

ords_prods_merged.loc[ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# Set Regular_customer flag

ords_prods_merged.loc[(ords_prods_merged['max_order'] <= 40) & (ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# Set New_customer flag

ords_prods_merged.loc[ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# check output

ords_prods_merged['loyalty_flag'].value_counts(dropna=False)
```

```python
ords_prods_merged[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

# 04. Compare Spending Habits by Loyalty Level

```python
ords_prods_merged.head(5)
```

```python
ords_prods_merged.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

# 05. Create Spending Flag Based on Average Price per User

```python
# group ords_prods_merged by user_id and find mean prices

ords_prods_merged['avg_spending'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# check output

ords_prods_merged.head(5)
```

```python
# Set customer spending flag for high spenders

ords_prods_merged.loc[ords_prods_merged['avg_spending'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Set customer spending flag for low spenders

ords_prods_merged.loc[ords_prods_merged['avg_spending'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# check output 

ords_prods_merged[['user_id', 'spending_flag', 'order_number']].head(15)
```

```python
# check output

ords_prods_merged['spending_flag'].value_counts(dropna=False)
```
