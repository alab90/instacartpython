Next, do consistency checks for missing values, duplicate values, mixed-type data.

```python
# checking for mixed types

for col in customers.columns.tolist():
  weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)
```

There are mixed data types in the first_name column. This might be becuase there are null values. 

```python
# checking for null values

customers.isnull().sum()
```

```python
# look for null values

customer_null = customers[customers['first_name'].isnull() == True]
```

```python
customer_null
```

There are 11,259 missing values in the first_name column. Our analysis will not required full names of customers, especially since we have unique user IDs. Also, from a data security perspective, it's probably best to remove identifying information. Therefore, we will create a version dropping first_name and last_name.

```python
# remove first_name and last_name

customers_clean = customers.drop(['first_name','last_name'],axis=1)
```

```python
# check output

customers_clean.head()
```

New customers_clean dataset has just user id with no first or last names

```python
# checking for duplicates

customers_clean_dups = customers_clean[customers_clean.duplicated()]
```

```python
customers_clean_dups
```

No duplicates found

# 03. Combine customer with ords_prods data

```python
# check output

customers_clean.head()
```

```python
# check output

df_ords_prods_merged_v4.head()
```

Both customers_clean and df_ords_prods_merged_v4 datasets share the user_id variable. Merge based on this key column.

```python
# Merge customers_clean & df_ords_prods_merged_v4

ords_prods_cust_merged = df_ords_prods_merged_v4.merge(customers_clean, on=['user_id'])
```

```python
# check output

ords_prods_cust_merged.head()
```

customers_clean and ords_prods_merged are now combined into one file based on the user_id key column
